```python
import plotly.plotly as py
from plotly.graph_objs import *

trace_2001 = Bar(x=df.CountryCode,
                  y=df.D2001,
                  name='2001',
                  marker=dict(color='#ffcdd2'))

trace_2017 = Bar(x=df.CountryCode,
                y=df.D2017,
                name='2017',
                marker=dict(color='#A2D5F2'))

trace_2019 = Bar(x=df.CountryCode,
                y=df.D2019,
                name='2019',
                marker=dict(color='#59666D'))

data = [trace_2001, trace_2017, trace_2019]


layout = Layout(title="GDP Comparision for 2001 , 2017, 2019",
                xaxis=dict(title='Country'),
                yaxis=dict(title='Value'))

plotly.offline.plot({
    "data": data,
    "layout": layout
})

#fig = Figure(data=data, layout=layout)

#py.iplot(fig, sharing='secret', filename='jupyter/styled_bar')
```
